```
import numpy as np
import pandas as pd
```

```
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa','profilre_score','lpa'])
df
```

```
import tensorflow
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.layers import Dense, Input
```

```
model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(2, activation='linear'))
model.add(Dense(1, activation='linear'))
```

```
model.summary()
```

```
model.get_weights()
```

```
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```
model.fit(df.iloc[:,0:-1].values,df['lpa'].values,epochs=200,verbose=1,batch_size=1)
```

```
model.get_weights()
```

```
new_data = np.array([[6,6]])

prediction = model.predict(new_data)

print("Predicted LPA:", prediction[0][0])
```
